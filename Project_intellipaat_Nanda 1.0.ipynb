```python
5. How manytimes does the price of Dubai oil exceed the price of Brent oil by a certain
 threshold $10 ?
```

```python
sum_of_dubai_oil_prices = df['oil_dubai'].sum()
sum_of_brent_oil_prices = df['oil_brent'].sum()
print(sum_of_dubai_oil_prices)
print(sum_of_brent_oil_prices)
Difference_betwwen_dubai_and_brent=(sum_of_dubai_oil_prices - sum_of_brent_oil_prices)
print(Difference_betwwen_dubai_and_brent)
threshold=(Difference_betwwen_dubai_and_brent > 10).sum()
print(threshold )
```

OBSERVATION:-


*   It is approxiamately same
*   the price of Dubai oil exceed the price of Brent oil by a certain threshold $10 is 0 Times.

 7. Which commodity experienced the highest price fluctuations during the observed period?


```python
Fluctationsofprice=df[['oil_brent', 'oil_dubai', 'coffee_arabica',
       'coffee_robustas', 'tea_columbo', 'tea_kolkata', 'tea_mombasa',
       'sugar_eu', 'sugar_us', 'sugar_world']].std()
sns.barplot(y=Fluctationsofprice.index,x=Fluctationsofprice.values,orient='h')
plt.title('Fluctations of price')
plt.show()
```

OBSERVATION:-
The oil brent has the highest fluctuations.     
The oil dubai has the 2nd highest fuctuations.

*  The oil brent has the highest fluctuations.
*   The oil dubai has the 2nd highest fuctuations.



But,the oil Brent and oil dubai is near small difference.

```python
 8. How has brent oil prices vary on a quarterly basis since the last five years?
```

```python
df['date'] = pd.to_datetime(df['date'])
df['Year'] = df['date'].dt.year
last_five_years_data = df[df['Year'] >= 2018]
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='oil_brent', data=last_five_years_data, marker='o', label='oil_brent', color='gray')
plt.title(' The last five years of oil brent')
plt.show()
```
